```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set global style
sns.set_theme(style="whitegrid")   # nice background style
plt.rcParams['figure.figsize'] = (10,6)  # default figure size
plt.rcParams['axes.titlesize'] = 16      # title font size
plt.rcParams['axes.labelsize'] = 12      # label font size
```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding = 'latin-1')
```

```python
data.head()
```

# Let’s start by looking at the descriptive statistics of the dataset

```python
data.describe()
```

```python
data.info()
```

# Converting Date Columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

# Adding New Date-Based Columns

```python
data['Order Month']=data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Day of Week']=data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly Sales Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=sales_by_month, x='Order Month', y='Sales', marker='o')
plt.title("Monthly Sales Analysis", fontsize=16)
plt.xlabel("Order Month")
plt.ylabel("Sales")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
sales_by_month
```

# Sales Analysis by Category

```python
import numpy as np

# Prepare data
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()

# Calculate percentages and absolute values for labels
total = sales_by_category['Sales'].sum()
labels = [
    f"{cat}\n{val/total:.1%}\n(${val:,.0f})" 
    for cat, val in zip(sales_by_category['Category'], sales_by_category['Sales'])
]

# Explode slices slightly for emphasis
explode = [0.05] * len(sales_by_category)

# Create pie chart
plt.figure(figsize=(8,8))
plt.pie(
    sales_by_category['Sales'], 
    labels=labels,
    autopct='%1.1f%%',              # we already built labels
    startangle=140,
    explode=explode,
    colors=sns.color_palette("Set2"),  # nice contrasting palette
    shadow=True
)

plt.title("Sales Analysis by Category", fontsize=18, fontweight='bold')
plt.show()
```

```python
sales_by_category
```

# Sales Analysis by Sub-Category

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=sales_by_subcategory, x='Sub-Category', y='Sales', palette='viridis')
plt.title("Sales Analysis by Sub-Category", fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_subcategory
```

# Sales Analysis by Region

```python
sales_by_region = data.groupby('Region')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=True)

plt.figure(figsize=(10,6))
plt.barh(sales_by_region['Region'], sales_by_region['Sales'], color=sns.color_palette("muted"))
plt.title("Sales by Region", fontsize=16)
plt.xlabel("Sales")
plt.ylabel("Region")
plt.show()
```

```python
sales_by_region
```

# Sales vs Profit by Category

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Sales', y='Profit', hue='Category', alpha=0.7, palette='Set2')
plt.title("Sales vs Profit by Category", fontsize=16)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.legend(title="Category")
plt.grid(True, alpha=0.3)
plt.show()
```

# Monthly Profit Analysis

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=profit_by_month, x='Order Month', y='Profit', marker='o', linewidth=2.5, color='teal')

# Annotate data points
for x, y in zip(profit_by_month['Order Month'], profit_by_month['Profit']):
    plt.text(x, y+0.02*profit_by_month['Profit'].max(), f"{y:,.0f}", 
             ha='center', va='bottom', fontsize=9, color='black')

plt.title("Monthly Profit Analysis", fontsize=18, fontweight='bold')
plt.xlabel("Order Month")
plt.ylabel("Profit")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
profit_by_month
```

# Sales and Profit Analysis by Customer Segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Melt data for seaborn grouped barplot
df_melt = sales_profit_by_segment.melt(id_vars='Segment', value_vars=['Sales', 'Profit'], 
                                       var_name='Metric', value_name='Amount')

plt.figure(figsize=(10,6))
sns.barplot(data=df_melt, x='Segment', y='Amount', hue='Metric', palette='pastel')

# Add value labels on bars
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width()/2, 
             p.get_height() + 0.01*df_melt['Amount'].max(), 
             f"{int(p.get_height()):,}", 
             ha='center', va='bottom', fontsize=9)

plt.title("Sales and Profit Analysis by Customer Segment", fontsize=18, fontweight="bold")
plt.xlabel("Customer Segment")
plt.ylabel("Amount")
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```

# analyse sales-to-profit ratio

```python
import numpy as np

sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = np.where(
    sales_profit_by_segment['Profit'] != 0,
    sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit'],
    np.nan  # or you can use 0 or some placeholder
)

print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
